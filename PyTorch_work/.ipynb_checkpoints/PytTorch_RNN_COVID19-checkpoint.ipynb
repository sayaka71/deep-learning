```python
import torch
import torch.nn as nn
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline

import datetime
from pandas.plotting import register_matplotlib_converters
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('time_series_covid19_confirmed_global.csv')
df.head()
```

```python
df = df[df['Country/Region'] == 'Japan']
df.head()
```

```python
df = df.iloc[:, 4:].copy()

daily_japan = df.iloc[0,:]
daily_japan.index = pd.to_datetime(daily_japan.index)
daily_japan
```

```python
sns.set()
plt.figure(figsize=(12,5), dpi=100)
plt.plot(daily_japan)
plt.title('COVID-19 in Japan')
plt.ylabel('Person injected')
plt.show()
```

```python
y = daily_japan.values.astype(float)
```

```python
def sequence_creator(input_data, window):
    """
    一個ずつずらして学習データを作る関数
    window: 区間
    """
    dataset = []
    data_len = len(input_data)
    for i in range(data_len - window):
        window_fr = input_data[i:i+window]
        label = input_data[i+window:i+window+1]
        dataset.append((window_fr, label))
        return dataset
```

```python
class LSTM_COVID(nn.Module):
    # 入力の次元サイズ，隠れそうのサイズ，出力のサイズ（翌日の感染者数の予測）
    def __init__(self, in_size=1, h_size=30, out_size=1):
        super().__init__()
        
        self.h_size = h_size
        # pytorchのモデル
        self.lstm = nn.LSTM(in_size, h_size)
        # 全結合層
        self.fc = nn.Linear(h_size, out_size)
        
        self.hidden = (torch.zeros(1,1,self.h_size), torch.zeros(1,1,self.h_size))
    
    # 予測する時のメソッド
    def forward(self, sequence_data):
        lstm_out, self.hidden = self.lstm(sequence_data.view(len(sequence_data), 1, -1), self.hidden)
        pred = self.fc(lstm_out.view(len(sequence_data), -1))
        return pred[-1]
                 
```

```python
torch.manual_seed(3)
model = LSTM_COVID()
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
epochs = 100
window_size = 7
loss_list = []
loss_temp = 0

# model.train()
```

```python
# 正規化．-1~1の間の数になるようにしている
scaler = MinMaxScaler(feature_range=(-1,1))
y_normarized = scaler.fit_transform(y.reshape(-1,1))
y_normarized = torch.FloatTensor(y_normarized).view(-1)
full_data = sequence_creator(y_normarized, window_size)
```

```python
for epoch in range(epochs):
    for sequence_in, y_train in full_data:
        
        optimizer.zero_grad()
        model.hidden = (torch.zeros(1,1,model.h_size), torch.zeros(1,1,model.h_size))
        
        y_pred = model(sequence_in)
        
        loss = criterion(y_pred, y_train)
        loss_temp += loss
        loss.backward()
        optimizer.step()
        
    if ((epoch+1) % 10 == 0):
        loss_list.append(loss_temp.item()/(10*len(full_data)))
        print(f'Epoch: {epoch+1} \nLoss: {loss_temp.item()/(10*len(full_data))}')
        loss_temp = 0
```

```python
plt.plot(loss_list)
plt.title('Loss')
```

```python
upcoming_future = 14
predictions = y_normarized[-window_size:].tolist()
model.eval()

for i in range(upcoming_future):
    sequence = torch.FloatTensor(predictions[-window_size:])
    with torch.no_grad():
        model.hidden = (torch.zeros(1,1,model.h_size), torch.zeros(1,1,model.h_size))
        predictions.append(model(sequence).item())
        
predictions_y = scaler.inverse_transform(np.array(predictions[:7]).reshape(-1,1))
x = np.arange('2020-06-02', '2020-06-09', dtype='datetime64[D]').astype('datetime64[D]')
```

```python
plt.figure(figsize=(12,4), dpi=100)
plt.title('COVID-19 Japan')
plt.plot(daily_japan)
plt.plot(x, predictions_y)
```
